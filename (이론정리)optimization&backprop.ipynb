Neural Network 개괄

linear score function: f=Wx
neural network는 functions의 nesting이다.
layer의 개수에 따라 살펴보면,
2 layers: f = W2max(0,W1x)
3 layers: f = W3max(0,W2max(0,W1x))
4 layers: f = W4tanh(W3, max(0,W2max(0,W1x)))
5 layers: f = W5*sigmoid(W4tanh(W3, max(0,W2max(0,W1x))))

input layer -> hidden layer 1 -> hidden layer 2 -> .. -> output layer
오른쪽으로 갈수록 depth, 위로 갈수록 (각 layer에서의 개수) width

gradients는 왜 중요할까?

딥러닝에서 W같은 parameter을 update할 때는 gradient descent같은 최적화 기법을 씀.
이때 loss function을 각 parameter에 대해 미분한 기울기를 계산해줘야 함.

-> backpropagation(역전파)가지고 output layer부터 input layer까지 gradient를 단계적으로 전파시키면서 계산함 (chain rule)
